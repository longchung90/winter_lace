### Business trending
***

* Although there are still missing values for us to handle, I am curious to know how the business activity was trending over that period of time using a data visualisation tool like a line graph. To do this, the method `pd.to_datetime(df['InvoiceDate']` allows coversion of values in column InvoiceDates into the datetime format of **yyyy-mm-dd**, which makes it easier to extract the month component needed for the line chart using the `dt.to_period('M')` method.

* It would be now possible to group the number of invoices and then count them by the months using the `groupby()` and `count()` functions.

* To be safe, I would do the group for the negative and the positive separarely.


#### 1. Separate the postive and negative into two different data_frames  and group the invoices by months

```python
# Convert the Invoice Date column to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create separate DataFrames for positive and negative quantities using .loc[] (for Pandas newer version)
positive_df = df.loc[df['Quantity'] > 0]
negative_df = df.loc[df['Quantity'] < 0]

# Extract Month and Year
positive_df['Month'] = positive_df['InvoiceDate'].dt.to_period('M')
negative_df['Month'] = negative_df['InvoiceDate'].dt.to_period('M')

# Group by Month and count the number of invoices
count_positive = positive_df.groupby('Month')['InvoiceNo'].count().reset_index(name='Count')
count_negative = negative_df.groupby('Month')['InvoiceNo'].count().reset_index(name='Count')

# Display the count DataFrames
print("Count of Positive Invoices per Month:")
print(count_positive)

print("\nCount of Negative Invoices per Month:")
print(count_negative)
```

#### 2. Using a line chart to quick-look at the trends if any and disover what months show a sale skyrocket.

```python
# Merge the two DataFrames on Month
merged_counts = pd.merge(count_positive, count_negative, on='Month', suffixes=('_Positive', '_Negative'))

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(merged_counts['Month'].astype(str), merged_counts['Count_Positive'], marker='o', label='Positive Quantities')
plt.plot(merged_counts['Month'].astype(str), merged_counts['Count_Negative'], marker='o', label='Negative Quantities')

# Adding titles and labels
plt.title('Monthly performance from Dec 2010 to Dec 2011')
plt.xlabel('Month')
plt.ylabel('Count of Invoices')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.figtext(0.5, 0.01, 'Figure 1: Raw data showing the business trending in sales by months', ha='center', fontsize=10)
# Show the plot
plt.tight_layout()
plt.show()
```

#### Trends:
* The above figure tells us that the business activity went very well and steady. The blue line representing the total number of invoices generated for the sale activities occuring on the retail website. 
* Despite a slow start at the begining of the year, their performance became very steady and strongly stablised half way though the year.
* In Sep, the performance went skyrocked and peaked in Nov 2011 and dropped back down again in Dec that year.
* Overall, the retail performance was at its peak, given there are negative invoices to consider but as in this case even if those negative points later offset the positive, a strong trend like this is hard to change.
#### Explanations:
* The slow start at the beginning is a normal phenomenon since it was  straight  after Chrismas, which was the biggest shopping time. Most people had  already done the shopping.
* In Mar onwards, with Easter holidays, the sales started picking up in a slow and steady fashion.

Figure 2 shows boxplots but not really boxplot due to those datapoints that have very high values. So, I applied the following method to calculate the quartiles and hence, removed identified outliers according and redid the boxplots.

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1_quantity = df_cleaned['Quantity'].quantile(0.25)
Q3_quantity = df_cleaned['Quantity'].quantile(0.75)
IQR_quantity = Q3_quantity - Q1_quantity

Q1_unitprice = df_cleaned['UnitPrice'].quantile(0.25)
Q3_unitprice = df_cleaned['UnitPrice'].quantile(0.75)
IQR_unitprice = Q3_unitprice - Q1_unitprice

# Determine outlier thresholds
lower_bound_quantity = Q1_quantity - 1.5 * IQR_quantity
upper_bound_quantity = Q3_quantity + 1.5 * IQR_quantity

lower_bound_unitprice = Q1_unitprice - 1.5 * IQR_unitprice
upper_bound_unitprice = Q3_unitprice + 1.5 * IQR_unitprice

# Remove outliers
df_cleaned_no_outliers = df_cleaned[
    (df_cleaned['Quantity'] >= lower_bound_quantity) & (df_cleaned['Quantity'] <= upper_bound_quantity) &
    (df_cleaned['UnitPrice'] >= lower_bound_unitprice) & (df_cleaned['UnitPrice'] <= upper_bound_unitprice)
]

# Display the shape of the original and cleaned DataFrame
print(f"Original DataFrame shape: {df_cleaned.shape}")
print(f"After removing the outliers, the updated DataFrame shape: {df_cleaned_no_outliers.shape}")
```

```python
import pandas as pd

# Sample DataFrame with invoice dates and product details
# df_cleaned_no_outliers should already contain your cleaned data

# Step 1: Extract month from invoice dates
df_cleaned_no_outliers['InvoiceDate'] = pd.to_datetime(df_cleaned_no_outliers['InvoiceDate'])
df_cleaned_no_outliers['Month'] = df_cleaned_no_outliers['InvoiceDate'].dt.month

# Step 2: Define a function to map months to seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'

# Step 3: Apply the function to create a new column for seasons
df_cleaned_no_outliers['Season'] = df_cleaned_no_outliers['Month'].apply(get_season)

# Step 4: Group by season and product to count the number of sales
seasonal_product_sales = df_cleaned_no_outliers.groupby(['Season', 'Description']).size().reset_index(name='SalesCount')

# Display the results
print(seasonal_product_sales)
```
