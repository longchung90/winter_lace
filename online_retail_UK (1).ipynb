#### Business trending

* From what we know so far regarding the dataset: it contains a large volumne of the transactions from the retail's business activites over the time period from Dec 2010 to Dec 2011. 

* Although there are still missing values for us to handle, I am curious to know how the business activity was trending over that period of time using a data visualisation tool like a line graph. To do this, the method `pd.to_datetime(df['InvoiceDate']` allows coversion of values in column InvoiceDates inton the datetime format of **yyyy-mm-dd**, which makes it easier to extract the month component needed for the line chart using the `dt.to_period('M')` method.

* It would be now possible to group the number of invoices and then count them by the months using the `groupby()` and `count()` functions.

* To be safe, I would do the group for the negative and the positive separarely.


```python
# Convert the Invoice Date column to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create separate DataFrames for positive and negative quantities using .loc[]
positive_df = df.loc[df['Quantity'] > 0]
negative_df = df.loc[df['Quantity'] < 0]

# Extract Month and Year
positive_df['Month'] = positive_df['InvoiceDate'].dt.to_period('M')
negative_df['Month'] = negative_df['InvoiceDate'].dt.to_period('M')

# Group by Month and count the number of invoices
count_positive = positive_df.groupby('Month')['InvoiceNo'].count().reset_index(name='Count')
count_negative = negative_df.groupby('Month')['InvoiceNo'].count().reset_index(name='Count')

# Display the count DataFrames
print("Count of Positive Invoices per Month:")
print(count_positive)

print("\nCount of Negative Invoices per Month:")
print(count_negative)
```

* Using a line chart to show the trends and the month of the best performance would be also revealed

```python
# Merge the two DataFrames on Month
merged_counts = pd.merge(count_positive, count_negative, on='Month', suffixes=('_Positive', '_Negative'))

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(merged_counts['Month'].astype(str), merged_counts['Count_Positive'], marker='o', label='Positive Quantities')
plt.plot(merged_counts['Month'].astype(str), merged_counts['Count_Negative'], marker='o', label='Negative Quantities')

# Adding titles and labels
plt.title('Monthly performance from Dec 2010 to Dec 2011')
plt.xlabel('Month')
plt.ylabel('Count of Invoices')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
def calculate_outlier_bounds(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

# Example for 'TotalSales'
lower_bound, upper_bound = calculate_outlier_bounds(df_cleaned, 'TotalSales')
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Identify outliers
outliers = df_cleaned[(df_cleaned['TotalSales'] < lower_bound) | (df_cleaned['TotalSales'] > upper_bound)]
print(f"Number of Outliers: {len(outliers)}")
```

```python
# Remove outliers from the DataFrame based on TotalSales
df_cleaned_no_outliers = df_cleaned[(df_cleaned['TotalSales'] >= lower_bound) & (df_cleaned['TotalSales'] <= upper_bound)]

# Check the number of rows before and after removing outliers
print(f"Original DataFrame Size: {df_cleaned.shape[0]}")
print(f"DataFrame Size After Removing Outliers: {df_cleaned_no_outliers.shape[0]}")
```
